```python
import pandas as pd

try:
    data = pd.read_csv("stocks4.csv", encoding='utf-8')
except UnicodeDecodeError:
    data = pd.read_csv("stocks4.csv", encoding='gbk')
data.head()
```

```python
# 显示数据的基本信息，包括行数、列数、列的数据类型等
print("数据基本信息：")
print(data.info())
```
